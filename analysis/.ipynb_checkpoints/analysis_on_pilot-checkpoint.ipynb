```python
def model_time_base(delay,alpha,sigma):
    modelPerceivedError=[alpha*(x/sigma)*math.exp(-(x**2)/sigma**2) for x in delay]
    modelPerceivedError=[j*1000 for j in modelPerceivedError]
    return modelPerceivedError
```

```python
path="C:/Users/omeru/Documents/Omer_Repos/Internship Pascal/Expectation_Shapes_Perceived_Time_final/analysis/data/"
data_Erva="niho_pilot_base_expectation_shapes_perceived_time_2023-03-16_23h46.19.804.csv" # no expectation
data_Nicola=  'nicola_base_pilot_expectation_shapes_perceived_time_2023-03-20_15h03.16.305.csv'# no expectation
data_Omer="299271_expectation_shapes_perceived_time_2023-03-19_15h42.48.633.csv" # no expectation
#data
data_Omer2="omer_pilot2_expectation_shapes_perceived_time_2023-03-17_00h51.58.576.csv" # expectation
df=pd.read_csv(path+data_Omer2) #
df=df[df["isTrial"]=="trial"]
#df=df.iloc[40:]
df["delayS"]=df["delay"]/1000
df["pTemporalError"]=df["perceivedTime"]-df["target.started"]
df["pTemporalErrorMS"]=df["pTemporalError"]*1000
df["physicalDelay"]=df["distractor.started"]-df["target.started"]
df["physicalDelayMS"]=df["physicalDelay"]*1000
df=df[df["pTemporalError"].notna()]
df.head()
```

```python
#normalize
#df["pTemporalErrorNormalized"] =2*(df["pTemporalError"] - min(df["pTemporalError"])) / ( max(df["pTemporalError"]) - min(df["pTemporalError"]))-1
avgError=np.mean(df["pTemporalError"])
df["pTemporalErrorAvaraged"]=df["pTemporalError"]-avgError
df_means=df.groupby('delayS').mean()
df_stdDev=df.groupby("delayS").std()
a=list(df["delay"].unique())
a.sort()
df_means["delayMs"]=a
df_means.head()
#df_stdDev
```

```python
xdata=df_means["physicalDelay"]
ydata=df_means["pTemporalErrorAvaraged"]
parameters, covariance = curve_fit(model_time_base, xdata, ydata)
```

```python
fit_A = parameters[0]
fit_B = parameters[1]
print(fit_A)
print(fit_B)
middleInt =[j for j in range(-350,350,1)]

middleInt=[j/1000 for j in middleInt]

fit_y = model_time_base(middleInt, fit_A, fit_B)
plt.figure(figsize=(15, 12))
plt.plot(xdata, ydata, 'o', label='data')

plt.plot(middleInt, fit_y, '-', label='fit')
plt.axvline(x=0, color='black', linestyle='dotted')
plt.axhline(y=0, color='black', linestyle='dotted')
#plt.set
#plt.yticks(list(np.linspace(-1,1,10)))
plt.legend()
```
